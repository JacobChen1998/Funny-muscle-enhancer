```python
import keras
from tensorflow.keras.datasets import mnist
from tensorflow.keras.models import Sequential
from tensorflow.keras.utils import Sequence

from tensorflow.keras.layers import Dense, Activation, Flatten, Input
from tensorflow.keras.layers import Conv2D, MaxPooling2D, UpSampling2D, BatchNormalization, Conv2DTranspose
import matplotlib.pyplot as plt
from keras import backend as K
import numpy as np
from keras.preprocessing.image import array_to_img, img_to_array, load_img
from sklearn.utils import shuffle
import os
import cv2
from tensorflow.keras.callbacks import Callback
from tensorflow.keras.models import Model
from tensorflow.keras.models import load_model
from cv2 import COLOR_LAB2RGB
```

```python
model = load_model('model_175_epochs.h5')
```

```python
x = cv2.imread('==your sample image ==')
x = cv2.resize(x,(224,224))
```

```python
def process(x):
    x =x/255
    pred_data = []
    pred_data.append(x)
    pred_data = np.array(pred_data)
    y = model.predict(pred_data)
    y = np.uint8(y[0]*255)
    return y
```

```python
for i in range(5):
    if i == 0:
        y = process(x)
        plt.imshow(cv2.cvtColor(y, cv2.COLOR_BGR2RGB))
        plt.show()
    else:
        y =  process(y)
        plt.imshow(cv2.cvtColor(y, cv2.COLOR_BGR2RGB))
        plt.show()
```
