```python
import cv2
import SkinDetector.skin_detector.skin_detector as skin_detector
from SkinDetection.SkinDetection import SkinDetection
import os
import numpy as np
import matplotlib.pyplot as plt
from Pedestrian_Detection_OpenCV.Human_Detection import Detector
import imutils
```

```python
def skin_nonskin_sep(img_org,skin_part):
    processed_img = np.zeros(img_org.shape).astype(int)
    non_processed_img = 255*np.ones(img_org.shape).astype(int)

    for x in range(img_org.shape[0]):
        for y in range(img_org.shape[1]):
            if skin_part[x][y] == 255:
                processed_img[x][y] = img_org[x][y]
            else:
                non_processed_img[x][y] = img_org[x][y]
    processed_img = np.uint8(processed_img) 
    non_processed_img = np.uint8(non_processed_img) 
    return processed_img, non_processed_img
```

```python
def contrast(img):
    brightness = 10;contrast = -15
    img = np.int16(img)
    img = img * (contrast/127+1) - contrast + brightness
    img = np.clip(img, 0, 255)
#     img = cv2.GaussianBlur(img, (3, 3), 0)

    img = np.uint8(img)
    return img
```

```python
def customed_white_to_black(img):
    img_gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    
    for x in range(img.shape[0]):
        for y in range(img.shape[1]):
            if img_gray[x][y] == img_gray[0][0]:
                img[x][y] = np.array([0,0,0])
    return img
```

```python
def main_process(fname):
    img = cv2.imread(fname)
    img_org = img.copy()

    skin_part = 255 - SkinDetection(img_org)
    skin_part2 = skin_detector.process(img_org)
    skin_part += skin_part2 
    skin_part = 255 - skin_part

    non_processed_img,processed_img = skin_nonskin_sep(img_org,skin_part)

    reverse_img = contrast(processed_img)
    reverse_img = customed_white_to_black(reverse_img)
    org_remove = customed_white_to_black(processed_img)
    result = reverse_img + non_processed_img
    rm_bg_result = reverse_img
    # plt.imshow(cv2.cvtColor(skin_part, cv2.COLOR_BGR2RGB))
    # plt.show()
    # plt.imshow(cv2.cvtColor(non_processed_img, cv2.COLOR_BGR2RGB))
    # plt.show()
    # plt.imshow(cv2.cvtColor(processed_img, cv2.COLOR_BGR2RGB))
    # plt.show()
    # plt.imshow(cv2.cvtColor(reverse_img, cv2.COLOR_BGR2RGB))
    # plt.show()
    # plt.imshow(cv2.cvtColor(img_org, cv2.COLOR_BGR2RGB))
    # plt.show()
    # plt.imshow(cv2.cvtColor(result, cv2.COLOR_BGR2RGB))
    # plt.show()
    return result,img_org,org_remove,rm_bg_result
```

```python
org_path = 'muscle_dataset/'
before = 'before_filtering/'
after = 'after_filtering/'
before2 = 'before_filtering_rm_bg/'
after2  = 'after_filtering_rm_bg/'
fnames = os.listdir(org_path)
fnames
```

```python
i = 0

for fname in fnames:
    
    print('org name: ', fname)
    result,img_org,org_remove,rm_bg_result = main_process(org_path+fname)
    new_fname = '%04d.png'%i
    print('new name: ', new_fname)
    cv2.imwrite(before+new_fname,result)
    cv2.imwrite(after+new_fname,img_org)
    cv2.imwrite(before2+new_fname,rm_bg_result)
    cv2.imwrite(after2+new_fname,org_remove)
    i += 1
```
